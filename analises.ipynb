```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot
import pandas as pd
```

```python
game = "CartPole-v0"
train_loss = pd.read_csv(game+"/train_loss.csv",delimiter=",",header=None)
train_scores = pd.read_csv(game+"/train_scores.csv",delimiter=",",header=None)
test_scores = pd.read_csv(game+"/test_scores.csv",delimiter=",",header=None)
size = (4,4)
```

```python
p = train_loss.plot(figsize=size)
p = train_loss.hist(figsize=size,range=(train_loss.min()[0],train_loss.max()[0]))
```

```python
p = train_scores.plot(figsize=size)
p = train_scores.hist(figsize=size,range=(train_scores.min()[0],train_scores.max()[0]))
```

```python
print("Pontuao mdia:",test_scores.mean()[0])
p = test_scores.plot(figsize=size)
p = test_scores.hist(figsize=size,range=(test_scores.min()[0],test_scores.max()[0]))
```
